# Decision Tree Regressor

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('datasets/decision_tree_regressor_data1.csv')
df2 = pd.read_csv('datasets/decision_tree_regressor_data2.csv')
```

# Let's look at data 1 

```python
df.columns
```

```python
X = df[['Outlook', 'Temperature', 'Humidity', 'Windy']]
```

```python
y = df['Hours Played']
```

```python
X
```

```python
y
```

# Decision Tree

```python
# Importing model
from sklearn.tree import DecisionTreeRegressor
```

```python
# Initialising Model
clf = DecisionTreeRegressor()
```

```python
# Fitting model
clf = clf.fit(X, y)
```

# We have to encode !!

### One Hot Encoding

```python
X_encoded = pd.get_dummies(X)
```

```python
# Initialising Model
clf = DecisionTreeRegressor(min_samples_split=3)
```

```python
# Fitting model
clf = clf.fit(X_encoded, y)
```

```python
# My Model has been trained
clf
```

# Plotting the decision tree

```python
from sklearn import tree
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,12))
tree.plot_tree(clf)
plt.show()
```
