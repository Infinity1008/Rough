#### Question 5
Assume, you are a trader and have invested over the years for your retirement. Lately, with the fluctuating economical condition, you are worried about the average return on investment. What average method is appropriate to compute the averagereturn for the data given in the table?

### Answer 5

```python
# geometric mean is suitable to calculate the average return. 

trader = pd.DataFrame({"Year":[2015,2016,2017,2018,2019,2020],
                        "Return":[0.36,0.23,-0.48,-0.30,0.15,0.31],
                        "Asset Price": [5000,6400,7890,9023,4567,3890],})
trader['gmean'] = trader['Return']+1
gmean = trader['gmean'].prod()
average_return = gmean**(1/len(trader)) -1
print("{:.2%}".format(average_return))
```

#### Question 6

Suppose you have been told to measure the height of all men on the earth. What strategy would you use?

### Answer

As we know that it is difficult to calculate the height of all men on the earth so we will calculate the height of all men on 
the earth by sampling variance method. In this method we collect the some data of men's height through the survey then, we will
use sampling method. for. e.g.
i have collected some data of men's height through the survey men_height = 155,165,169,170,172,165,150,155 then we can used the
sample variance which help me to identify the men's height on the earth.

```python
men_height = [155,165,169,170,172,165,150,155]
men_height_1 = np.array(men_height)
men_height_1
```

```python
mean_men_height = men_height_1.mean()
s_variance_men = (men_height_1 - mean_men_height)**2
s_variance_men_height = (s_variance_men).sum()/(len(men_height_1) -1)
s_std_men_height = (s_variance_men_height)**0.5
print('variance:',s_variance_men_height)
print('standard deviation:',s_std_men_height)
```
