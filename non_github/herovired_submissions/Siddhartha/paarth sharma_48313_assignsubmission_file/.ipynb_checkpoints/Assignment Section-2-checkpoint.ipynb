To observe similarity in distribution between age & experience, we'll calculate correlation.

The correlation between age & experience is 0.99 which suggests a very strong positive linear relationship between
the two. It shows that as age increases, experience tends to increase.This correlation coefficient suggests that there is a very consistent pattern where individuals with higher ages tend to have higher levels of experience.

# Q20

```python
X = bank["Income"]
m = bank["Income"].mean()
print("mean =", m)
std = bank["Income"].std()
print("std deviation =", std)
for i in range(0,len(X)):
    z = (X[i] - m)/std
    print(f"z-score for {X[i]} is", z)
```
