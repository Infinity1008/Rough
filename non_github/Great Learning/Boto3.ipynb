### Making the API Call

```python
response = client.detect_document_text(
    Document={
        'Bytes': b
        }
)
```

### Converting and Saving to JSON

```python
import json
json_obj = json.dumps(response) 
```

```python
# Writing to sample.json
with open("response_detect.json", "w") as outfile:
    outfile.write(json_obj)
```

```python
response_for_forms
```

```python
json.dumps(response_for_forms)
```

```python
for i in response_for_forms['Blocks']:
    if i['BlockType'] == 'KEY_VALUE_SET':
        print(i['Relationships'])
```

```python
for i in response_for_forms['Blocks']:
    if i['BlockType'] == 'WORD':
        if i['Id'] == '0b56db39-1152-494c-a5d7-1d81660cd2a6':
            print(i)
```

```python
for i in response_for_forms['Blocks']:
    if i['BlockType'] == 'WORD':
        print(i)
```
