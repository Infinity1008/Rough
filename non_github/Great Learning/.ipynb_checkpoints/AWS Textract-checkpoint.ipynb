### Analysing Document using bytes

```python
response_from_b = client.analyze_document(
    Document={
        'Bytes': b,
    },
    FeatureTypes=[
        'FORMS'
    ]
)
```

```python
import json

json.dumps(response_from_b)
```

```python
for i in response_from_b['Blocks']:
    if i['BlockType'] == 'KEY_VALUE_SET':
        print(i)
```
