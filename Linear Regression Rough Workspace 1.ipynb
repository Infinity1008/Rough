```python
import numpy as np
```

```python


x = np.array([np.random.randint(0,100) for i in range(100)])
y = np.array([np.random.randint(0,100) for i in range(100)])

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)

ax.axline((0, 0), slope=0.5, color='C0', label='by slope')


plt.scatter(x,y) # Plotting data
# plt.xticks(list(range(compute_till)), list(range(compute_till))) # Redefining x-axis labels

fig.set_figwidth(10)
fig.set_figheight(7)

plt.title('Random Data')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
