<center><h1>Poisson Distribution</h1></center>

## Poisson Process
A Poisson Process is a model for a series of discrete event where the average time between events is known, but the exact timing of events is random. The arrival of an event is independent of the event before (waiting time between events is memoryless). For example, suppose we own a website which our content delivery network (CDN) tells us goes down on average once per 60 days, but one failure doesn’t affect the probability of the next. All we know is the average time between failures. This is a Poisson process that looks like:

![poisson](https://miro.medium.com/max/720/1*YzGP8RRR9gYa6f8pJGIgMg.png)

The important point is we know the average time between events but they are randomly spaced (stochastic). We might have back-to-back failures, but we could also go years between failures due to the randomness of the process.

## Poisson Distribution
The Poisson Process is the model we use for describing randomly occurring events and by itself, isn’t that useful. We need the Poisson Distribution to do interesting things like finding the probability of a number of events in a time period or finding the probability of waiting some time until the next event.

The Poisson Distribution probability mass function gives the probability of observing k events in a time period given the length of the period and the average events per time:

# Formula
For the sake of information, the formula for poisson distribution is:- 


$$\left( x \right) = \frac{{e^{ - \lambda } \lambda ^x }}{{x!}}$$


We will not go in depth about the derivation of this formula, just know that we will use ```scipy``` to calculate this for us, so we don't have to worry about that.   

# Using Scipy
We can use scipy to calculate the probability for us. 

### Example
A store sells 3 apples per day on average. What is the probability that they will sell 5 apples on a given day? 

- ```mu``` is where we put our average rate of events  
- ```k``` is where we put the expected number of events we want to calculate the probability for.

````python
from scipy.stats import poisson
poisson.pmf(k=5, mu=3)
````

<center>If you want, you can directly use the formula above to calculate this yourself.</center>

This is not recommended, just use ```Scipy``` instead

````python
# Writing the formula in code
import numpy as np
from scipy.special import factorial

def calc_prob(events, minutes, k):
    # Calculate probability of k events in specified number of minutes
    lam = events * minutes
    return np.exp(-lam) * np.power(lam, k) / factorial(k)
````

- 3 events, so 3 in events
- per day, so 1
- and we want to calculate the probability of 5 events happening in the day, so 5

````python
calc_prob(3,1,5)
````

Same result, just calculated directly.

# Let's Visualize
So we have learned couple of things, 
- We need the rate at which the events are happening, per minute or per hour or whatever you wish
- After we have this rate, we can calculate the probability of a particular number of events happening
- So we can calculate the probability of 1 event happening, or 2 or 3 and so of. 

So let's just plot this for a better understanding. We will use the same example as above
- A store sells 3 apples per day on average. What is the probability that they will sell 5 apples on a given day?

The rate we have in this is example is 3, and we want to see the probability for 1,2,3,4 sales in a day let's say.

**Answer**:-

I am defining a function that plots a poisson proababiltiy distribution, you don't have to understand this right now, just focus on the plot

````python
import matplotlib.pyplot as plt

def create_poisson_plot(lamda, compute_till):
    poisson_result = []
    for i in range(compute_till):
        poisson_result.append(poisson.pmf(k=i, mu=lamda))
    
    fig = plt.figure()
    ax = fig.add_subplot(111)

    plt.plot(range(compute_till),poisson_result,marker='o') # Plotting data
    plt.xticks(list(range(compute_till)), list(range(compute_till))) # Redefining x-axis labels

    fig.set_figwidth(10)
    fig.set_figheight(7)

    plt.title('Poisson')
    plt.xlabel('Number of occurence')
    plt.ylabel('Probability of occurence')


    prob_of_lamda_round = [round(i,ndigits=3) for i in poisson_result]

    for i, v in enumerate(prob_of_lamda_round):
        ax.annotate(str(v), xy=(i,v), xytext=(-7,7), textcoords='offset points')

#     plt.ylim(0,0.2)
    plt.show()
````

````python
create_poisson_plot(3,10)
````

- With this plot, we can see the probability of seeing 0 sale, 1 sale,2,3,4 and so on. 
### This is the power of poisson, we can see the probability of a number of events happening given we have a rate.

I have written some code, that creates the plots for you given the rate, don't mind the code part, just focus on the plot and play around.

````python
from ipywidgets import interact, interactive
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')


def create_poisson_plot(lamda, compute_till):
    poisson_result = []
    for i in range(compute_till):
        poisson_result.append(poisson.pmf(k=i, mu=lamda))
    
    fig = plt.figure()
    ax = fig.add_subplot(111)

    plt.plot(range(compute_till),poisson_result,marker='o') # Plotting data
    plt.xticks(list(range(compute_till)), list(range(compute_till))) # Redefining x-axis labels

    fig.set_figwidth(10)
    fig.set_figheight(7)

    plt.title('Poisson')
    plt.xlabel('Number of occurence')
    plt.ylabel('Probability of occurence')


    prob_of_lamda_round = [round(i,ndigits=3) for i in poisson_result]

    for i, v in enumerate(prob_of_lamda_round):
        ax.annotate(str(v), xy=(i,v), xytext=(-7,7), textcoords='offset points')

#     plt.ylim(0,0.2)
    plt.show()

    
lamda_value = widgets.FloatSlider(
    value=3,
    min=0,
    max=10,
    step=0.1,
    description='lamda (rate)',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
#     readout_format='d',
    style={'description_width': 'initial'}
)

till = widgets.IntSlider(
    value=10,
    min=0,
    max=15,
    base=10,
    description='Compute till',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
    style={'description_width': 'initial'}
)


interactive_plot = interactive(create_poisson_plot,lamda=lamda_value,compute_till=till)
interactive_plot
````

## Question 3
What is the probability of seeing 4 stars or less in a hour

````python
# You can add up the probabilities for seeing 0,1,2,3,4. Let's do just that
prob_of_seeing_0 = poisson.pmf(k=0, mu=5)
prob_of_seeing_1 = poisson.pmf(k=1, mu=5)
prob_of_seeing_2 = poisson.pmf(k=2, mu=5)
prob_of_seeing_3 = poisson.pmf(k=3, mu=5)
prob_of_seeing_4 = poisson.pmf(k=4, mu=5)

print(f'\nThe probability of seeing 0 stars is {prob_of_seeing_0}')
print(f'\nThe probability of seeing 1 stars is {prob_of_seeing_1}')
print(f'\nThe probability of seeing 2 stars is {prob_of_seeing_2}')
print(f'\nThe probability of seeing 3 stars is {prob_of_seeing_3}')
print(f'\nThe probability of seeing 4 stars is {prob_of_seeing_4}')
````

````python
print('\n-------------------- Adding them together --------------------\n')

sum_of_all = prob_of_seeing_0 + prob_of_seeing_1 + prob_of_seeing_2 + prob_of_seeing_3 + prob_of_seeing_4

print(f'Summing all together, the result is {sum_of_all}')
````

So our answer is:-  44% is the probability of seeing 4 stars or less.

We can similarly create plots for all the other probabilities as well, but we want all the curves in one plot, for that i'll write some custom code, again, don't focus on the code part for now, focus on the resulting plots.

````python
import matplotlib.pyplot as plt

prob_of_star4 = [calc_prob((4/60),60,i) for i in range(15)]
prob_of_star8 = [calc_prob((8/60),60,i) for i in range(15)]
prob_of_star12 = [calc_prob((12/60),60,i) for i in range(15)]
prob_of_star16 = [calc_prob((16/60),60,i) for i in range(15)]

range_of_stars = [i for i in range(15)]
prob_of_star = [prob_of_star4, prob_of_star8, prob_of_star12, prob_of_star16]
````

````python
fig = plt.figure()
ax = fig.add_subplot(111)

fig.set_figwidth(10)
fig.set_figheight(7)


labels_x = ['4/60','8/60','12/60','16/60']

for k,i in enumerate(prob_of_star):
    plt.plot(range(len(range_of_stars)),i,marker='o',label=labels_x[k]) # Plotting data
#     plt.xticks(range(len(range_of_stars)), range_of_stars) # Redefining x-axis labels


plt.title('Stars')
plt.xlabel('Number of stars')
plt.ylabel('Probability of stars')

for j in prob_of_star:
    prob_of_star_round = [round(i,ndigits=2) for i in j]

    for i, v in enumerate(prob_of_star_round):
        ax.annotate(str(v), xy=(i,v), xytext=(-7,7), textcoords='offset points')

plt.ylim(0,0.2)
plt.legend(loc="upper right")
plt.show()
````
