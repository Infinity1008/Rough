#### Question 6
Which country players have the highest wages on average. (Hint: Column name 'WageEUR')

```python
df.groupby('Nationality')[['WageEUR']].mean().sort_values(by='WageEUR')
```

#### Question 7
Which club has on average the most expensive players (Hint: Column name 'ValueEUR')

```python
df.groupby('Club')['ValueEUR'].mean().sort_values()
```

#### Question 8
Which is the best club (No hints! Figure it out ( TT_TT), Have fun (^_^ )

```python
df.groupby('Club')['Overall'].mean().sort_values().tail(10)
```
