```python
from sklearn.impute import SimpleImputer

imp1 = SimpleImputer(missing_values = np.nan, strategy = 'median')
df1['displacement'] = imp1.fit_transform(df1[['displacement']])
df1['horsepower'] = imp1.fit_transform(df1[['horsepower']])
df1['weight'] = imp1.fit_transform(df1[['weight']])

imp2 = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
df1['origin'] = imp2.fit_transform(df1[['origin']])
```

```python
df2 = df1.dropna()
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.info()
```

```python
df2['cylinders'] = df2.cylinders.astype('int64')
q1 = df2.quantile(0.25)
q3 = df2.quantile(0.75)
IQR = q3 - q1
lower_limit , upper_limit = q1 - 1.5*IQR , q3 + 1.5*IQR
df3 = df2[~((df2<lower_limit) |(df2>upper_limit)).any(axis = 1)]
df3.shape
```

```python
def vif_cal(input_data, dependent_col):
    x_vars = input_data.drop([dependent_col], axis = 1)
    xvar_names = x_vars.columns
    for i in range(0, xvar_names.shape[0]):
        y = x_vars[xvar_names[i]]
        x = x_vars[xvar_names.drop(xvar_names[i])]
        rsq = smf.ols(formula = 'y~x', data = x_vars).fit().rsquared
        vif = round(1/(1-rsq), 2)
        print(xvar_names[i], "VIF = ", vif)
```

```python
df3_num = df3.drop(['origin', 'cylinders','weight', 'horsepower'], axis = 1)
vif_cal(df3_num, 'mpg')
```

```python
from sklearn.model_selection import train_test_split
test_size = 0.2
seed = 10
train_set , test_set = train_test_split(df3, test_size = test_size, random_state = seed)
```

```python
train_set.shape, test_set.shape
```

```python
model = smf.ols(formula = 'mpg ~ origin + cylinders + displacement + acceleration + Age_of_Vehicle', data = train_set)
results = model.fit()
print(results.summary())
```

```python
model1 = smf.ols(formula = 'mpg ~ cylinders + weight + Age_of_Vehicle', data = train_set)
results1 = model1.fit()
print(results1.summary())
```

```python
residuals = results1.resid
Train_RMSE = np.sqrt(np.mean(residuals**2))
Train_RMSE
```

```python
y_pred = results1.predict(test_set[['cylinders', 'weight', 'Age_of_Vehicle']])
```

```python
Test_RMSE = np.sqrt(np.mean((test_set['mpg'] - y_pred)**2))
```

```python
Test_RMSE
```
