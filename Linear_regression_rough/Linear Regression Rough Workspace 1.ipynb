```python
import numpy as np
```

```python
random_slope = np.random.normal()
random_intercept = np.random.normal()
line_points = np.array([(random_slope*i + random_intercept) for i in range(0,100,2)])

random_noise = np.array([np.random.randint(-100,100) for i in range(0,100,2)])
data_points = line_points * random_noise


x = np.array(range(0,100,2))
x = x.reshape(-1,1)


# ----------------------------------------------------------
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)

# ax.axline((0, 0), slope=0.5, color='C0', label='by slope')


plt.scatter(x,data_points) # Plotting data
# plt.xticks(list(range(compute_till)), list(range(compute_till))) # Redefining x-axis labels

fig.set_figwidth(10)
fig.set_figheight(7)

plt.title('Random Data')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(x, data_points)

# The coefficients
print("Coefficients: \n", regr.coef_)

# The intercept
print("Intercept: \n", regr.intercept_)

# ------------------------------------------------
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)

# ax.axline((0, 0), slope=0.5, color='C0', label='by slope')
abline_values = [regr.coef_ * i + regr.intercept_ for i in x]

plt.scatter(x,data_points) # Plotting data
# plt.xticks(list(range(compute_till)), list(range(compute_till))) # Redefining x-axis labels

fig.set_figwidth(10)
fig.set_figheight(7)

plt.title('Random Data')
plt.xlabel('X')
plt.ylabel('Y')

plt.plot(x, abline_values, '--')

plt.show()
```

```python
def create_data_and_line_plot_residuals():
    random_slope = np.random.normal()
    random_intercept = np.random.normal()

    our_range = [i for i in range(0,100,2)]

    line_points = np.array([(random_slope*i + random_intercept) for i in our_range])
    random_noise = np.array([np.random.randint(-100,100) for i in our_range])
    data_points = line_points + random_noise

    x = np.array(range(0,100,2))
    x_reshaped = x.reshape(-1,1)
    
    # ----------------------------------------------------------

    import matplotlib.pyplot as plt
    fig = plt.figure()
    ax = fig.add_subplot(111)

    plt.scatter(x_reshaped,data_points) # Plotting data

    fig.set_figwidth(10)
    fig.set_figheight(7)

    # ----------------------------------------------------------

    from sklearn import linear_model
    from sklearn.metrics import mean_squared_error, r2_score

    # Create linear regression object
    regr = linear_model.LinearRegression()

    # Train the model using the training sets
    regr.fit(x_reshaped, data_points)

    # ax.axline((0, 0), slope=0.5, color='C0', label='by slope')
    abline_values = [regr.coef_ * i + regr.intercept_ for i in x_reshaped]
    plt.plot(x_reshaped, abline_values, '--')

    # ----------------------------------------------------------

    # Plotting the residual values
    import random
    random_choice = [random.randint(0,49) for i in range(6)]
    random_choice

    for i in random_choice:
        plt.plot([x[i],x[i]], [abline_values[i][0], data_points[i]], '--')

    # ----------------------------------------------------------
        
    predicted_values = np.array([i[0] for i in abline_values])   
    ssr = np.sum(np.square(data_points - predicted_values))

#     plt.text(.05, .05, f'The SSR is {ssr}', fontsize = 10)

    # ----------------------------------------------------------

    plt.title(f'Generating Random Data, fitting a line to it. Plotting the residuals\n--------------------------\n The residual is {ssr}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
interact_manual(create_data_and_line_plot_residuals,i=widgets.Button(description="Click Me!"));
```

```python
from IPython.display import display

random_slope = np.random.normal()
random_intercept = np.random.normal()

our_range = [i for i in range(0,100,2)]

line_points = np.array([(random_slope*i + random_intercept) for i in our_range])
random_noise = np.array([np.random.randint(-100,100) for i in our_range])
data_points = line_points + random_noise


# ----------------------------------------------------------


# ----------------------------------------------------------

# Plotting the residual values
import random
random_choice = [random.randint(0,49) for i in range(10)]
random_choice


# ----------------------------------------------------------

import matplotlib.pyplot as plt

def draw_line(slope, intercept):    
    global data_points
    global random_choice
    x = np.array(range(0,100,2))
#     x_reshaped = x.reshape(-1,1)
    
    fig = plt.figure()
    ax = fig.add_subplot(111)

    plt.scatter(x,data_points) # Plotting data

    fig.set_figwidth(10)
    fig.set_figheight(7)

    
    """Plot a line from slope and intercept"""
#     x = np.linspace(0,100,50)
    abline_values = slope * x + intercept
    
    # Plot the best fit line over the actual values    
    plt.xlim([-5, 100])
    plt.ylim([-5,100])
    plt.plot(x, abline_values, '--')
    plt.title(slope)
    
    # Plotting residuals
    for i in random_choice:
        plt.plot([x[i],x[i]], [abline_values[i], data_points[i]], '--')


    # Plotting SSR
    predicted_values = np.array(abline_values)
    ssr = np.sum(np.square(data_points - predicted_values))
    
    
    plt.title(f'Generating Random Data, fitting a line to it. Plotting the residuals\n--------------------------\n The sum of squared residual is {ssr}')
    plt.show()

w = interactive(draw_line, 
                slope = widgets.FloatSlider(min=-10, max=10, step=0.1, value=1),
               intercept = widgets.FloatSlider(min=-5, max=10, step=0.5, value=0))
```
