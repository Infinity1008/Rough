# Let's save the outputs of these

````python
# The first song
temp1 = []
for i in playlist.items():
    if 'SONGID01' in i[1]:
        print(i[0])
        temp1.append(i[0])
        
# The second song
temp2 = []
for i in playlist.items():
    if 'SONGID06' in i[1]:
        print(i[0])
        temp2.append(i[0])
        
# The third song
temp3 = []
for i in playlist.items():
    if 'SONGID16' in i[1]:
        print(i[0])
        temp3.append(i[0])
````

# Now, we want to see which user has most common with new user

We can do this by appending all the lists and just seeing which user has the most count.

````python
# Let's add all 3 lists together
common_users = temp1 + temp2 + temp3
````

````python
common_users
````

````python
common_users.count('USER65')
````

````python
for i in common_users:
    print(common_users.count(i))
````

# We can tidy that up a bit

````python
for i in common_users:
    print({i:common_users.count(i)})
````

# You know what, we can even save it in a dataframe

````python
import pandas as pd
users = []
common_counts = []

for i in common_users:
    users.append(i)
    common_counts.append(common_users.count(i))

common_df = pd.DataFrame({'Users':users, 'Common Songs Count':common_counts})
````

````python
common_df
````

# We can now sort it 

````python
common_df.sort_values(by='Common Songs Count',ascending=False)
````
