### Question 2
Plot the total cases for India over the months.

```python
df[['date','total_cases']]
```

```python
df['date']
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['week'] = df['date'].dt.week
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
df_india = df[df['location'] == 'India']
```

```python
df_time = df_india[['date','week','month','year','total_cases']]
```

```python
df_time.head()
```

```python
df_time.tail()
```

```python
sns.lineplot(data=df_time[df_time['year'] == 2020], x="month", y="total_cases")
```

```python
sns.lineplot(data=df_time[df_time['year'] == 2021], x="month", y="total_cases")
```

```python
sns.lineplot(data=df_time[df_time['year'] == 2022], x="month", y="total_cases")
```
