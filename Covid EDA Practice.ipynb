### Now, hear me out, I might be wrong.
- What if, in the first row we had a smoker and in the 2nd there was not. 
- It is repeating the 1st row value in 2nd until the value of the next row actually changes. Meaning it would keep repeating the first value until there is a legit smoker in row 4th or nth and so on.
- So, essentially all those repeated values are 0, and only the first value is ```TRUE```. 
- Again, this is a hypotheses, it could work or not.
- If it does work, then the previous question will need to be re-answered.
- But, how will we know if this is the case or not, we have no one to tell us if that's the case or not.
- Well, here we rely on our data, if we make all the repeated values 0, and then do an analysis
 - If our analysis comes out to be sensible, we might have just hit the jackpot with our hypotheses.
 - If our analysis makes no sense, there we had nothing to begin with anyway, so in no case we can lose, a win-win in my books.

# Testing our hypotheses.
Well, how do we do this.
- How do we make all the repeated values 0
- Well, first of all we will need to figure out when the value changes
 - The way we can do that is we subtract row 2 from row 1 and if the answer is 0, the value is the same.
     - If it does change than that subtraction would lead to a non zero number.
     - So, i would subtract all the rows from the rows before it, and then search for all the rows which did **not** give me a 0.
     - Then i can get an index for all the rows where things actually changed and where they did not.
     - Then i can replace all the same values with 0 and let the rows where things changed to unchanged. 

### Well, that is the plan. 
![](https://i.imgflip.com/751mc9.jpg)

````python
# Our column
df['female_smokers']
````

````python
# Our column shifted down 1 row
df['female_smokers'].shift(1)
````

````python
# Essentiall, subtracting row 2 from row 1 and so on for all the rows
df['female_smokers'] - df['female_smokers'].shift(1)
````

See how we have rows with 0, exactly what we wanted. 
- Anywhere we have 0's we know the values did not change
- But where are the rows where the value did change, do they even exist.
- Let's plot to confirm

````python
# Plotting our differences
(df['female_smokers'] - df['female_smokers'].shift(1)).plot()
````

# It works
![](https://i.imgflip.com/751mx7.jpg)

We can see here, all the rows that were the same resulted in a zero and all the rows when there was a change, we see a positive or a negative number. 

So now
- We can make all the rows that are 0, and change their value from repeated to 0.

````python
(df['female_smokers'] - df['female_smokers'].shift(1))
````

````python
# First thing first, let's make these NAN's also 0
(df['female_smokers'] - df['female_smokers'].shift(1)).fillna(0)
````

````python
# Perfect, now we can check for values equal to 0
(df['female_smokers'] - df['female_smokers'].shift(1)).fillna(0) == 0
````

````python
# Now we need to take all these TRUE's and use it to change their orignal value to 0
repititive_array = (df['female_smokers'] - df['female_smokers'].shift(1)).fillna(0) == 0
````

````python
# Let's use this True/False array to get the rows which are repetitive
df[repititive_array]
````

````python
# Looking at our column
df[repititive_array]['female_smokers']
````

So, in all these rows, there is a repitition. Let's change that

````python
# Changing the value
df[repititive_array]['female_smokers'] = 0
````
